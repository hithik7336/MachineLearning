# MULTIPLE LINEAR REGRESSION 

## Importing The Libraries & Dataset

```python
# IMPORTING THE LIBRARIES

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline 

import seaborn as sns
```

```python
# IMPORTING THE DATASET

data = pd.read_csv('50_Startups.csv')  # GETTING OUR DATASET

X = data.iloc[:,:4]    # GETTING THE INDEPENDENT VARIABLES
y = data.iloc[:,4]     # GETTING THE DEPENDENT VARIABLES
```

```python
# CHECKING THE HEAD OF DATA

data.head()
```

```python
# GETTING OUR DEPENDENT VARIABLE

y
```

```python
# GETTING OUR INDEPENDENT VARIABLES

X
```

## Data Preprocessing

```python
# CREATING THE DUMMY VARIABLES OF THE STATE COLUMN

dummies = pd.get_dummies(X['State'])
```

```python
# GETTING OUR DUMMy VARIABLES

dummies
```

```python
# CONCATENATING THE DUMMY VARIABLES

X = pd.concat([X, dummies], axis = 1)
```

```python
# DROPPING THE STATE COLUMN

X = X.drop('State', axis = 1)
```

```python
# CHECKING THE X AGAIN

X
```

## Splitting Data To Train & Test Set

```python
# IMPORTING THE train_test_split

from sklearn.model_selection import train_test_split
```

```python
# SPLITTING THE DATA

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Creating The Model, Training & Making Predictions


```python
# IMPORTING OUR LINEAR REGRESSION MODEL AND CREATING OBJECT OF MODEL

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
# TRAINING OUR MODEL ON TRAINING SET

regressor.fit(X_train, y_train)
```

```python
# GETTING THE PREDICTIONS FROM THE MODEL

y_pred = regressor.predict(X_test)
```

```python
# PREDICTIONS VECTOR

np.set_printoptions(precision = 2) # TO GET DECIMAL VALUES UPTO 2 PLACES

y_pred  
```

```python
# COMPARING THE PREDICTIONS WITH REAL RESULTS

np.concatenate((y_pred.reshape(len(y_pred),1), np.array(y_test).reshape(len(y_test),1)), axis = 1)
```

## Visualising The Results 

```python
# PLOTTING THE y_test & y_pred

plt.figure(figsize = (6,6))

sns.scatterplot(y_test,y_pred)
```
